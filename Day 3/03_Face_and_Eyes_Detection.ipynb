### Import Libraries

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifiers

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eyes_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### Open Image

```python
img = cv2.imread('Images/eye_face.jpg')
plt.imshow(img)
                 
```

### Fix Image

```python
fix_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

### Face Classifier

```python
faces = face_classifier.detectMultiScale(fix_image,1.3,5)
```

### If No Face Detected

```python
if faces is ():
    print('NO faces detected')
```

### It's a Kind of Magic

```python
def detect_face_eyes(fix_img) :
    face_rects = face_classifier.detectMultiScale(fix_image)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,(x,y),(x+w,y+h),(255,0,0),10)
        eyes_rects = eyes_classifier.detectMultiScale(fix_image)
        for (x,y,w,h) in eyes_rects:
            cv2.rectangle(fix_img,(x,y),(x+w,y+h),(0,0,255),10)    
    return fix_img
```

```python
result = detect_face_eyes(fix_image)
plt.imshow(result)
```
