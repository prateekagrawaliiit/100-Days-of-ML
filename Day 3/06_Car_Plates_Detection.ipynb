### Import Libraries

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 

%matplotlib inline
```

### Open Image

```python
img = cv2.imread('Images/car_plate.jpg')
```

### Convert Image

```python
def display(img) :
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

### Display it

```python
display(img)
```

### Create Classifier

```python
plates_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### It's a Kind of Magic

```python
def detect_plate(img) :
    plate_image = img.copy()
    
    plate_rects = plates_classifier.detectMultiScale(img,1.1,1)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),5)
        
    return plate_image
```

### Check the Results

```python
result = detect_plate(img)
plt.imshow(result)
```
